```python
# Writing a method that takes dataset, directory location, scale_variables and scale_variables_mapping and provides all the needed bar charts in respective folders

def create_bar_charts(dataset, dir_loc, scale_variables, scale_variables_mapping):

    # Create a directory to store the PNG files if it doesn't exist
    output_directory = os.path.join(current_path, dir_loc)
    os.makedirs(output_directory, exist_ok=True)

    # Iterate over each column and create a bar chart
    for column in scale_variables:
        # Map values to corresponding labels using varibles_mapping
        mapped_values = dataset[column].map(scale_variables_mapping)
        
        # Count occurrences of each value in the column
        value_counts = mapped_values.value_counts()
        
        # Calculate the total count
        total_count = value_counts.sum()
        
        # Calculate the percentage of total for each value
        percentages = value_counts / total_count * 100
        
        # Modify the column name to handle colons
        modified_column = column 
        
        # Get the title from the mapping dictionary
        title = column_name_mapping.get(column, column)  # Use column name if not found in mapping
        
        # Split the title into words and insert a newline after every seventh word
        words = title.split()
        words_with_newlines = [word + ('\n' if (i + 1) % 7 == 0 else '') for i, word in enumerate(words)]
        modified_title = ' '.join(words_with_newlines)
        
        # Plot the bar chart
        plt.figure(figsize=(10, 6))  # Adjust figure size if needed
        percentages.plot(kind='bar', color='skyblue')  # Plotting the bar chart with percentages
        
        # Annotate each bar with its percentage of total inside the bar near the top
        for i, v in enumerate(percentages):
            plt.text(i, v / 2, f'{v:.2f}%', ha='center', va='center', color='black', fontsize=10)  # Set font color to black
        
        plt.title(modified_title)  # Set title
        plt.ylabel('Percentage of Target Customers Surveyed')  # Set y-axis label
        plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
        plt.tight_layout()  # Adjust layout to prevent overlap
        
        # Save the plot as a PNG file
        # this following statement will replace the slash as creates problem in saving the file to required path
        modified_column = modified_column.replace("/","-")
        output_file = os.path.join(output_directory, f"{modified_column}_bar_chart.png")
        plt.savefig(output_file)
        
        # Close the plot to free memory
        plt.close()
```
